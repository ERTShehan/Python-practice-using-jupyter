```python
class Car:
    wheels = 4 #class attribute

    def drive_to(self, destination):
        print(f"Driving to {destination} on {self.wheels} wheels")

    def show_wheel_count(self):
        print(f"The car runs on {self.wheels} wheels")

my_car = Car()
my_car.drive_to("Matara")
my_car.show_wheel_count()
```

Write a python class name Circle with the flowing specification
* And initializer that take the cirlcle's radius as a parameter and stores.
* And instance method "area" that returns the area of the circle.
* And instance method circumference that returns the circumference of the circle
* And instance method "display_info" that print circle with radius <radius> has area <area> and circumference <circumference>

```python
from math import pi

class Circle:
    def __init__ (self, radius):
        self.radius = radius

    def area (self):
        return math.pi * (self.radius ** 2)
    
    def circumference(self):
        return 2 * math.pi * self.radius
    
    def display_info(self):
        area = self.area()
        circumference = self.circumference()
        print(f"Circle with radius {self.radius} has area {area} and circumference {circumference}")

my_circle = Circle(5)
my_circle.display_info()
```

<h3>OOP concepts in python</h3>

```python
# Inheritance

class Animal:
    def __init__ (self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} speak")

class Dog(Animal):
    def speak(self): #method overriding
        print(f"{self.name} says woof")

dog_1 = Dog("snoopy")
dog_1.speak()
```

```python
# Inheritance

class Animal: #superclass
    def __init__ (self, name):
        self.name = name
    
    def speak(self):
        print(f"{self.name} speak")

class Dog(Animal): #subclass
    def __init__ (self, name, color):
        # self.name = name
        super().__init__(name) #calling superclass constructor
        self.color = color
        

    def speak(self): #method overriding
        print(f"{self.name} says woof It is {self.color} color")

dog_1 = Dog("snoopy", "brown")
dog_1.speak()
```

Create a python class Shape with a name attribute and a method area that prints "Area calculate not defined for <"name">" then create two sub classes squre and circle that inherite from shape.Each sub class should,
* Have it's own constractor to initializer appropriate attribute
* override the area method to calculate and print area of the shape
* Use meaningful instance attribute for there calculation

```python
from math import pi

class Shape:
    def __init__ (self, name):
        self.name = name

    def area(self):
        print(f"Area calculate not defined for {self.name}")

class Square(Shape):
    def __init__ (self, name, side_length):
        super().__init__(name)
        self.side_length = side_length

    def area(self):
        area = self.side_length ** 2
        print(f"Area of {self.name} is {area}")

class Circle(Shape):
    def __init__ (self, name, radius):
        super().__init__(name)
        self.radius = radius

    def area(self):
        area = pi * (self.radius ** 2)
        print(f"Area of {self.name} is {area}")

square_1 = Square("MySquare", 4)
square_1.area()

circle_1 = Circle("MyCircle", 4)
circle_1.area()
```
