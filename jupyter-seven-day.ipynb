You are given flowing the pandas data frame containing product seles

df = pd.DataFrame({
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Headset"],
    "UnitSold": [15, 50, 34, 20, 40],
    "Price": [12000, 1500, 3500, 25000, 6000],
    "Discount": [0.10, 0.05, 0.08, 0.12, 0.07]
})

* Calculate the final price after discount for each product
* Calculate the total revenue for each product after discount
* For above part 1 and part 2 you should add both result as new coloms to the dataframe
* Filter and Display the the product where revenue is grater that 500,000
* Calculate the total revenue of all product combined

```python
import pandas as pd

df = pd.DataFrame({
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Headset"],
    "UnitSold": [12, 50, 34, 20, 40],
    "Price": [120000, 1500, 3500, 25000, 6000],
    "Discount": [0.10, 0.05, 0.08, 0.12, 0.07]
})

df["Final Price"] = df["Price"] - (df["Price"] * df["Discount"])
df["Total Revenue"] = df["Final Price"] * df["UnitSold"]
print(df)

filtered_df = df[df["Total Revenue"] > 500000]
print(filtered_df)

total_revenue = df["Total Revenue"].sum()
print(total_revenue)
```
