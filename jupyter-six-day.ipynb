# Polymorphism

```python
# Polymorphism

class Dog:
    def speak(self):
        return "Woof!"
class Cat:
    def speak(self):
        return "Meow!"
def animal_sound(animal):
    print(animal.speak())
dog = Dog()
cat = Cat()
animal_sound(dog)  # Output: Woof!
animal_sound(cat)  # Output: Meow!
```

# Encapsulation

```python
# Encapsulation

class square:
    def __init__(self, side_length):
        self.__side_length = side_length  # Private attribute

    def get_side_length(self):
        return self.__side_length  # Getter method

    def set_side_length(self, side_length):
        if side_length > 0:
            self.__side_length = side_length  # Setter method
        else:
            raise ValueError("Side length must be positive")

    def area(self):
        return self.__side_length ** 2  # Method to calculate area
    
sq = square(4)
print(sq.area())
```

```python
# private and protected attributes in Python

class Square:
    def __init__(self, name):
        self._name = name # Protected attribute

    def _area(self, side): # Protected method
        return side * side
    
sq = Square("MySquare")
area = sq._area(5) # Accessing protected method(Not recommended)
print(f"Square Name: {sq._name}, Area: {area}")
```

```python
# private attribute cannot be accessed directly

class Square:
    def __init__(self, name):
        self.__name = name # Private attribute

    def __area(self, side): # Private method
        return side * side
    
sq = Square("MySquare")
# area = sq.__area(5) # This will raise an AttributeError
# print(f"Square Name: {sq.__name}, Area: {area}")
```

```python
# use getter and setter methods to access private attributes and methods

class Square:
    def __init__ (self, name):
        self.__name = name # Private attribute

    def get_name(self):
        return self.__name  # Getter method

    def set_name(self, name):
        self.__name = name  # Setter method

    def __area(self, side_length): # Private method
        return side_length * side_length
    
    def get_area(self, side_length):
        return self.__area(side_length)
    
sq = Square("MySquare")
print(f"Square Name: {sq.get_name()}, Area: {sq.get_area(5)}")

sq.set_name("NewSquare")
print(f"Updated Square Name: {sq.get_name()}")
```
