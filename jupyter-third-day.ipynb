```python
# Structural Pattern Matching
x = 2
match x:
    case 1:
        print("x is 1")
    case 2:
        print("x is 2")
    case 3:
        print("x is 3")
    case _: # default case
        print("x is not 1, 2 or 3")
```

```python
# write a python program that takes days as input for week (monday, tuesday,...) and prints whether it is a weekday or weekend using structural pattern matching

day = input("Enter a day of the week: ")

match day:
    case "Saturday" | "Sunday": #Not work "or" in structural pattern
        print("Weekend days")
    case "Monday" | "Tuesday" | "Wednesday" | "Thursday" | "Friday":
        print("Weekdays")
    case _:
        print("Invalid day")
```

```python
# Matching by the 

numbers = [4, 3, 7]

match numbers:
    case[x, y]:
        print(x * y)
    case[x, y, z]:
        print(x + y + z)
    case _:
        print("This list does not contain 2 or 3 numbers")
```

```python
# Write a Python Program using Structural Pattern matching that takes a list of numbers
#  if the list Contain two numbers, read them as the base and hight of a right angle traingal and calculate the hipotenus
#  if the list contains three numbers, calculate the area of the traingal using those numbers as sides
#  otherwise, print "unsupported number of inputs"

numbers = [3, 4, 5]

match numbers:
    case[x, y]:
        hypotenuse = (x**2 + y**2) ** 0.5
        print("The hypotenuse is", hypotenuse)
    case[a, b, c]:
        s = (a + b + c) / 2
        area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
        print(area)
    case _:
        print("unsupported number of inputs")


# https://www.geeksforgeeks.org/dsa/c-program-find-area-triangle/
```

```python
# Calculate factorial number using for loop

number = int(input("Enter a number: "))

if number < 0:
    print("Negative number")
else:
    factorial = 1
    for i in range(1, number + 1):
        factorial *= i
    print("The factorial number of: ", factorial)
```

```python
# python program that takes 10 integers from the user using a for loop.
# * the program should stop taking input imediatly if the user enters a 0.
# * after the loop end, print the some of all the numbers entered before 0

total = 0
for i in range(10):
    num = int(input("Enter an integer"))
    if num == 0:
        break
    total += num
print("The sum of all numbers", total)
```
