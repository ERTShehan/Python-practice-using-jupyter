```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak (self):
        print(f"{self.name} speak")

class Dog(Animal):
    def __init__ (self, name, color):
        super().__init__(name)
        self.color = color

    def speak (self):
        print(f"{self.name} says woof {self.color} color")

dog_1 = Dog("Snovy", "Black")
dog_1.speak()
```

```python
class Dog():
    def speak (self):
        return "Woof"

class Cat():
    def speak (self):
        return "Meow"
    
def animal_sound(animal):
    print(animal.speak())

dog = Dog()
cat = Cat()
animal_sound(dog)
animal_sound(cat)
```
