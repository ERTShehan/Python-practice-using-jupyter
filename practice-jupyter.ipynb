```python
add = lambda x, y: x + y
print(add(3, 5))
```

```python
add = (lambda x, y: x + y)(3, 5)
print(add)
```

```python
max_number = lambda x, y: x if x > y else y
print(max_number(10, 20))
```

```python
numbers = [10, 15, 18, 20, 25, 30, 33, 40, 45, 51]
answers = list(filter(lambda x: x % 3 == 0 and x % 5 != 0, numbers))
print(answers)
```

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)
```

```python
from math import isqrt, log2

input_number = int(input("Enter a number: "))
print(isqrt(input_number))
print(log2(input_number))
```
