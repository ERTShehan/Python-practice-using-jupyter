```python
my_var = 5
print(my_var)
print(type(my_var))

# 2my_var = 10  # Invalid variable name, starts with a digit
# print(2my_var)

# my@var = 15  # Invalid variable name, contains special character '@'
# print(my@var)

MY_VAR = 20  # Valid, but not recommended (constant by convention)
print(MY_VAR)

myVar = 25  # Valid, but not recommended (camelCase)
print(myVar)

my_var_2 = 30  # Valid, recommended (snake_case)
print(my_var_2)

# my-var = 35  # Invalid variable name, contains hyphen '-'
# print(my-var)

PI_VALUE = 3.14159 # Constant use uppercase letters
```

```python
complex_num = 3 + 4j  # Complex number
print(complex_num)

a = 6.5e3  # Scientific notation for 6.5 * 10^3
print(a)

a = 5
b = float(a)  # Convert integer to float
print(b)

a = 5.99
b = int(a)  # Convert float to integer (truncates decimal part) 
print(b)
```

```python
my_list = ["Dog", "Cat", "Rabbit"]
my_list.remove("Cat") # Remove by value
print(my_list)


new_var = my_list.pop(1)
print(new_var)
print(my_list)

del my_list[0] # Delete by index
print(my_list)
```

```python
my_list = ["Dog", "Cat", "Rabbit"]
del my_list # Delete the entire list
# print(my_list) # This will raise an error since my_list is deleted

my_list = ["Dog", "Cat", "Rabbit"]
my_list.clear() # Clear all elements from the list
print(my_list) # This will print an empty list
```
