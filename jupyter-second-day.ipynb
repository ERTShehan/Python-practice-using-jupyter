```python
sports = ("Football") # This is not a tuple, it's a string
print(type(sports))
sports = ("Football",) # This is a tuple with one element
print(type(sports))
```

```python
# Python Dictionary
dictionary_1 = {"name":"suger",
                "weight": "1kg",
                "price": 175.00
               }
print(dictionary_1)
print(type(dictionary_1))

print(dictionary_1["name"]) # Accessing value using key

dictionary_1["price"] = 200.00 # Modifying value using key
print(dictionary_1)

dictionary_1["expiry"] = "2024-12-31" # Adding a new key-value pair
print(dictionary_1)

dictionary_1.pop("expiry") # Removing a key-value pair using key
print(dictionary_1)
```

```python
# Python Operators
# Arithmetic Operators: +, -, *, /, %, **, //

x = 5
y = 2
print(x + y) #Addtion
print(x - y) #Subtraction

print(x * y) #Multiplication
print(x ** y) # Exponentiation: 5 raised to the power of 2 = 25

print(x/2) #Division: 5 divided by 2 = 2.5
print(x//2) #Floor Division: 5 divided by 2 = 2

print(x % y) # Modulus: Remainder of the division of x by y = 1
```

```python
# Assignment Operators: =, +=, -=, *=, /=, %=, //=, **=
x = 5
print(x)

x += 2
print(x) # x = x + 2

x -= 2
print(x) # x = x - 2

x *= 2
print(x) # x = x * 2

x /= 2
print(x) # x = x / 2

x %= 2
print(x) # x = x % 2

x //= 2
print(x) # x = x // 2

x **= 2
print(x) # x = x ** 2
```

```python
# Comparison Operators: ==, !=, >, <, >=, <=
a = 5
b = 3
print(a == b) # Equal to
print(a != b) # Not equal to
print(a > b)  # Greater than
print(a < b)  # Less than
print(a >= b) # Greater than or equal to
print(a <= b) # Less than or equal to
```

```python
# Logical Operators: and, or, not
x = 5
y = 6
print(x > 1 and y < 10) # True and True = True
print(x > 1 and y > 10) # True and False = False

print("-------------------------------------")

print(x > 1 or y < 10)  # True or True = True
print(x > 1 or y > 10)  # True or False = True
print(x < 1 or y > 10)  # False or False = False

print("-------------------------------------")

print(not(x > 1 and y < 10)) # not True = False
print(not x > 1)
```

```python
# Identity Operators: is, is not
x = 5
y = 6

print(x is y) # False

x = 5
y = 5
print(x is y) # True


x = 257
y = 259
print(x is y) # False
y = 257
print(x is y) # False #Integer Interning in Python for small integers (-5 to 256)

print("-------------------------------------")

print(x is not y) # True
```

```python
# String Interning in Python
a = "hello"
b = "hello"
print(a is b) # True #No Limit for short strings without spaces

a = "hello world"
b = "hello world"
print(a is b) # False # Strings with spaces may not be interned
```

```python
# Membership Operators: in, not in
nums = [1, 6, 8, 11, 10]
print(6 in nums) # True
print(15 in nums) # False

print(6 not in nums) # False
print(15 not in nums) # True
```

```python
# Bitwise Operators: &, |, ^, ~, <<, >>
x = 5  # In binary: 0101
y = 10 # In binary: 1010
print(x & y)  # Bitwise AND: 0000 = 0

print(x | y)  # Bitwise OR: 1111 = 15
print(x ^ y)  # Bitwise XOR: 1111 = 15
print(~x)     # Bitwise NOT: 1010 = -6
print(x << 1) # Bitwise left shift: 1010 = 10
print(x >> 1) # Bitwise right shift: 0010 = 2
```

```python
# Operators Precedence
# exponentiation > multiplication/division/floor division/modulus > addition/subtraction > relational/comparison > equality > logical NOT > logical AND > logical OR > assignment
```
