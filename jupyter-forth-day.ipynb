```python
# Write a python function call square_sum to calculate the sum of square number from one to any given number n.

def square_sum(n):
    total = 0
    for i in range(1, n + 1):
        total += i ** 2
    return total

print(square_sum(1))
print(square_sum(2))
print(square_sum(3))
```

write a python function name add_corresponding_numbers that takes two list of numbers as input and return a new list containing the sum of corresponding elements from both lists

If the two list are not equal length, you should display and error message

```python
def add_corresponding_numbers(list1, list2):
    if len(list1) != len(list2):
        return "Lists are not of equal length"
    else:
        result = []
        for i in range(len(list1)):
            result.append(list1[i] + list2[i])
        return result
print(add_corresponding_numbers([1, 2, 3], [4, 5, 6]))
```

Write a python function name calculate_total that calculate and returns the total cost after a apply in a discount.
The function should
* take price as a requered arguments
* use quentity as and optional argument with a default value of one
* use discount as and optional arguments with a default value of zero
* apply the discount as a presentage on the total price

```python
def calculate_total(price, quantity=1, discount=0):
    total_price = price * quantity
    discount_amount = total_price * (discount / 100)
    final_price = total_price - discount_amount
    return final_price

print(calculate_total(100))
print(calculate_total(100, 3))
print(calculate_total(100, 3, 20))
```

<h2>Arbitrary Arguments</h2>

```python
# Arbitrary Positional Arguments

def arbitrary_positional_args(*args):
    print(args)
    print(type(args))

    for arg in args:
        print(arg)

arbitrary_positional_args(1, 5, 8)
arbitrary_positional_args('apple', 'banana')
```

write a python function name sum_numbers that accept any number of numeric arguments and return there sum

```python
def sum_numbers(*args):
    total = 0
    for num in args:
        total += num
    return total

print(sum_numbers(5, 10, 15))
print(sum_numbers(52,8,58,54))
```
