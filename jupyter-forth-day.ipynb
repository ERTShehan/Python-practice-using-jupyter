```python
# Write a python function call square_sum to calculate the sum of square number from one to any given number n.

def square_sum(n):
    total = 0
    for i in range(1, n + 1):
        total += i ** 2
    return total

print(square_sum(1))
print(square_sum(2))
print(square_sum(3))
```

write a python function name add_corresponding_numbers that takes two list of numbers as input and return a new list containing the sum of corresponding elements from both lists

If the two list are not equal length, you should display and error message

```python
def add_corresponding_numbers(list1, list2):
    if len(list1) != len(list2):
        return "Lists are not of equal length"
    else:
        result = []
        for i in range(len(list1)):
            result.append(list1[i] + list2[i])
        return result
print(add_corresponding_numbers([1, 2, 3], [4, 5, 6]))
```

Write a python function name calculate_total that calculate and returns the total cost after a apply in a discount.
The function should
* take price as a requered arguments
* use quentity as and optional argument with a default value of one
* use discount as and optional arguments with a default value of zero
* apply the discount as a presentage on the total price

```python
def calculate_total(price, quantity=1, discount=0):
    total_price = price * quantity
    discount_amount = total_price * (discount / 100)
    final_price = total_price - discount_amount
    return final_price

print(calculate_total(100))
print(calculate_total(100, 3))
print(calculate_total(100, 3, 20))
```

<h2>Arbitrary Arguments</h2>

```python
# Arbitrary Positional Arguments

def arbitrary_positional_args(*args):
    print(args)
    print(type(args))

    for arg in args:
        print(arg)

arbitrary_positional_args(1, 5, 8)
arbitrary_positional_args('apple', 'banana')
```

write a python function name sum_numbers that accept any number of numeric arguments and return there sum

```python
def sum_numbers(*args):
    total = 0
    for num in args:
        total += num
    return total

print(sum_numbers(5, 10, 15))
print(sum_numbers(52,8,58,54))
```

write a python function name average_positive that accept any number of numeric arguments. calculate the average of all positive number only. If no positive numbers are provided, the function should display no positive numbers provided.

```python
# 1st

def average_positive(*args):
    positive_numbers = [num for num in args if num > 0]
    print(positive_numbers)

    number_len = len(positive_numbers)
    if number_len == 0:
        return 0
    total = sum(positive_numbers)
    average = total / number_len
    return average

average_positive(10, -5, 1, -1, 4, 0)
```

```python
# 2nd type

def average_positive(*numbers):
    total = 0
    count = 0

    for num in numbers:
        if num > 0:
            total += num
            count += 1

    if count == 0:
        return "No positive number provided"
    else:
        return total/ count
        
print(average_positive(10, -5, 1, -1, 4))
average_positive(10, -5, 1, -1, 4, 0)
```

```python
# Arbitary Keyword Arguments

def arbitary_keyword_args(**kwargs):
    print(kwargs)
    print(type(kwargs))

    for key in kwargs.keys():
        print(key)

    for value in kwargs.values():
        print(value)

arbitary_keyword_args(name="Shehan", age=22, city="Matara")
```

```python
def arbitary_keyword_args(**kwargs):
    print(kwargs)
    print(type(kwargs))

    for key, value in kwargs.items():
        print(f"{key}: {value}")

arbitary_keyword_args(name="Shehan", age=22, city="Matara")
```

Write a python function name average_score that:
* Accept any number of keyword arguments where
    * keys are subject name(like math, physics)
    * Values are scores obtain in each subject

* Calculate and print the average score of all subject
* If no subject are provided, then print no subject provided

```python
def average_score(**kwargs):
    if len(kwargs) == 0:
        print("No subject provided")
        return

    total_score = sum(kwargs.values())
    number_of_subjects = len(kwargs)
    average = total_score / number_of_subjects
    print("Average Score:", average)

average_score(math=85, physics=90)
average_score(math=0, physics=50)
average_score()
```

# Build-in Function in Python

```python
# Absolute Value

def absolute_value(number):
    if number < 0:
        return -number
    else:
        return number
    
print(absolute_value(-5))
print(absolute_value(0))
print(absolute_value(7))

# abs() built-in function

print(abs(-5))
```

```python
# map() built-in function
def find_square(x):
    return x ** 2

list_1 = [6, 3, 4, 8]
result = map(find_square, list_1)
print(result)
print(type(result))

result_list = list(result)
print(result_list)
```

Write a function convert_to_celsius that convert a fahrenheit temperature to celsius, rounded to two decimal places. Then use the map function to convert the list [32, 68, 77] to celsius and print the result in list

celsius = (fahrenheit - 32) * 5/9

```python
def convert_to_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5 / 9
    return round(celsius, 2)

fahrenheit_list = [32, 68, 77]
celsius_list = list(map(convert_to_celsius, fahrenheit_list))
print(celsius_list)
```

```python
# filter() built-in function

numbers = [10, 9, 4, 8, 11]

def filter_even_numbers(num):
    return num % 2 == 0

even_numbers = filter(filter_even_numbers, numbers)
print(type(even_numbers))
print(list(even_numbers))
```

Use the filter function to extract all numbers from the list [12, 15, 18, 20, 25, 30, 35] that are divisible by three but not by five, and print the resulting list.

```python
number = [12, 15, 18, 20, 25, 30, 35]

def filter_divisible_by_3_not_5(num):
    return num % 3 == 0 and num % 5 != 0

resulting_numbers = filter(filter_divisible_by_3_not_5, number)
print(list(resulting_numbers))
```
